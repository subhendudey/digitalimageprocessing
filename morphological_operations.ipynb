```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#### Question

Consider the following rice image. There is a variation in the background illumination profile. The objective is to identify the object boundary. Write your algorithm to perform the task.

```python
image = cv2.imread("images/rice.jpg")
plt.imshow(image,"gray")
```

#### Answer

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
eq_img = cv2.equalizeHist(gray_img)
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(131)
plt.imshow(image, "gray")
plt.title("Original")
plt.subplot(132)
plt.imshow(gray_img, "gray")
plt.title("Grayscale")
plt.subplot(133)
plt.imshow(eq_img, "gray")
plt.title("Equalized")
```

#### Convert to binary

```python
binr_1 = cv2.threshold(gray_img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
binr_2 = cv2.threshold(eq_img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
```

```python
plt.figure(figsize=(8, 8))

plt.subplot(121)
plt.imshow(binr_1, "gray")
plt.title("Binary Image from source")
plt.subplot(122)
plt.imshow(binr_2, "gray")
plt.title("Binary Image from equalized")
```

#### Observation

The right bottom corner of the right image has greater binarization, whereas for other areas left image is good

#### Combine two binary image

```python
binr = np.zeros((935, 935), np.uint8)

for i in range(binr.shape[0]):
    for j in range(binr.shape[1]):
        if (i >=600 and j>= 600):
            binr[i][j] = binr_2[i][j]
        else:
            binr[i][j] = binr_1[i][j]
```

```python
plt.imshow(binr, "gray")
```

#### Define a kernel and perform erosion

```python
kernel = np.ones((5, 5), np.uint8)
erosion = cv2.erode(binr, kernel, iterations=1)
```

#### Subtracting the erosion from binary results in the edges

```python
edge = binr - erosion
```

```python
plt.imshow(edge, "gray")
```

#### Alternate ways

Morphological gradient first applies erosion and dilation individually on the image and then computes the difference between the eroded and dilated image. The output will be an outline of the given image.

```python
kernel = np.ones((3, 3), np.uint8)
morph_gradient = cv2.morphologyEx(binr,cv2.MORPH_GRADIENT, kernel)
```

```python
plt.imshow(morph_gradient, "gray")
plt.title("Morph_gradient Image")
```
