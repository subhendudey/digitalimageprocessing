# Basic Operations

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('Letter R.jpg')  
img.shape
```

```python
plt.figure(figsize=(3, 3))
plt.imshow(img)
```

### Covert to grayscale

```python
# Y' = 0.2989 R + 0.5870 G + 0.1140 B 

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
gray = rgb2gray(img)
```

```python
gray.shape
```

```python
plt.figure(figsize=(3, 3))
plt.imshow(gray, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.show()
```

### Transformation matrix for scale and rotation (color and grayscale)

```python
# create the scaling transformation matrix
T_s = np.array([[2, 0, 0], [0, 2, 0], [0, 0, 1]])
T_s_gray = np.array([[2, 0], [0, 2]])

# create the rotation transformation matrix
T_r = np.array([[0, 1, 0], [-1, 0, 0], [0, 0, 1]])
T_r_gray = np.array([[0, 1], [-1, 0]])

# create combined tranformation matrix
T = T_s @ T_r
T_gray = T_s_gray @ T_r_gray
```

```python
def binary_transform(img_as_array):
    t_image_as_array = np.empty((img_as_array.shape[0], img_as_array.shape[1]), dtype=np.uint8)
    for i, row in enumerate(t_image_as_array):
        for j, col in enumerate(row):
            pixel_data = t_image_as_array[i, j]
            if (pixel_data < 200):
                t_image_as_array[i, j] = 0
            else:
                t_image_as_array[i, j] = 255
    return t_image_as_array
```

```python
b_gray = binary_transform(gray)
```

```python
plt.figure(figsize=(3, 3))
plt.imshow(b_gray, cmap=plt.get_cmap('binary'), vmin=0, vmax=1)
plt.show()
```

### Rotation

```python
img_transformed = np.empty((1000, 1000), dtype=np.uint8)
for i, row in enumerate(gray):
    for j, col in enumerate(row):
        pixel_data = gray[i, j]
        input_coords = np.array([i, j])
        i_out, j_out = T_r_gray @ input_coords
        img_transformed[i_out, j_out] = pixel_data
```

```python
plt.figure(figsize=(3, 3))
plt.imshow(img_transformed, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.show()
```

### Scaling

```python
# 2x scaling requires a tranformation image array 2x the original image
img_transformed = np.empty((2000, 2000), dtype=np.uint8)
for i, row in enumerate(gray):
    for j, col in enumerate(row):
        pixel_data = gray[i, j]
        input_coords = np.array([i, j])
        i_out, j_out = T_s_gray @ input_coords
        img_transformed[i_out, j_out] = pixel_data
```

```python
plt.figure(figsize=(3, 3))
plt.imshow(img_transformed, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.show()
```
