```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#### Question

Consider the following coin image. Write a code which can count the number of coins present in it.

```python
image = cv2.imread("images/circles.jpg")
plt.imshow(image,"gray")
```

#### Answer

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
eq_img = cv2.equalizeHist(gray_img)
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(131)
plt.imshow(image, "gray")
plt.title("Original")
plt.subplot(132)
plt.imshow(gray_img, "gray")
plt.title("Grayscale")
plt.subplot(133)
plt.imshow(eq_img, "gray")
plt.title("Equalized")
```

#### Grayscale image is good enough - lets convert it to binary

```python
binr_img = cv2.threshold(gray_img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
```

#### Invert the image as the object segmentation / detection works better with while objects

```python
invert_img = cv2.bitwise_not(binr_img)
```

```python
plt.imshow(invert_img, "gray")
plt.title("Inverted")
```

#### Erode to separate the coins

```python
kernel = np.ones((35, 35), np.uint8)
erosion = cv2.erode(invert_img, kernel, iterations=1)
```

```python
plt.imshow(erosion, "gray")
plt.title("Eroded")
```

#### Apply image segmentation technique / connected component analysis to find the number of coins

```python
(cnt, hierarchy) = cv2.findContours(erosion.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
```

```python
print(len(cnt))
```

```python
analysis = cv2.connectedComponentsWithStats(erosion,4,cv2.CV_32S)
(totalLabels, label_ids, values, centroid) = analysis
```
